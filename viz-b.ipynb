```python
import seaborn as sns
sns.load_dataset('titanic')
```

```python
sns.set()
```

## The graph below shows the differences in age and sex when surviving the sinking of the Titanic. Based off this data, it appears more older females survived than older males.

```python
survived_titanic = sns.load_dataset("titanic")
sns.barplot(data=survived_titanic, x="sex", y="age", hue="survived")
```

## The line graph below is the most interesting because it compares the surviving rate based off both class and age. According to this data, few third class children survived but it looks like most of the second class and first class children did survive.

```python
sns.lineplot(data=survived_titanic, x="age", y="survived", hue="class")
```
